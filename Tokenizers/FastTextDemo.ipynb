# FastText Embeddings Demo

```python
import fasttext
```

Load the model

```python
vectorizer = fasttext.load_model('fasttexttokenizer.bin')
```

```python
phrase = input('Enter a sentence: ')
print(vectorizer.get_sentence_vector(phrase))
```

```python
phrase = input('Enter a sentence: ')
vectorizer.get_word_vector(phrase)
```
